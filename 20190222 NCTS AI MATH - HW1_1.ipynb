1. Import userpackage.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

2. Import dataset (Iris dataset from `scikit-learn`).

```python
from sklearn import datasets
iris = datasets.load_iris()
target = iris['target']
data = pd.DataFrame(iris['data'], columns=iris['feature_names'])
data.head(6)
```

> [Iris data](https://zh.wikipedia.org/wiki/%E5%AE%89%E5%BE%B7%E6%A3%AE%E9%B8%A2%E5%B0%BE%E8%8A%B1%E5%8D%89%E6%95%B0%E6%8D%AE%E9%9B%86)

3. Plot the scattor plot of iris data.

```python
headers = list(data.keys())
dim = len(headers)

n = 1
fig = plt.figure(figsize=(12, 12))
colors = ['red', 'green', 'blue']
for i in list(range(dim)):
    for j in list(range(dim)):
        ax = fig.add_subplot(dim, dim, n)
        if i != j:
            for index in list(range(3)):
                data.loc[target==index].plot.scatter(x=headers[j],y=headers[i],c=colors[index],\
                                                     label=list(iris['target_names'])[index],  \
                                                     ax=ax)
            ax.legend().set_visible(False)
        else:
            ax.text(0.25, 0.5, headers[i])
        n += 1
        if i == j:
            ax.xaxis.set_visible(False)
            ax.yaxis.set_visible(False)
        else:
            ax.xaxis.label.set_visible(False)
            ax.yaxis.label.set_visible(False)
        if i == 0 and j == 1:
            ax.legend(bbox_to_anchor=(3.5, 1), loc=2, borderaxespad=0., title="Species name:")
```
